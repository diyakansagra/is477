```python
import pandas as pd

df = pd.read_csv('merged_movies.csv')

# 1A. Average IMDb rating by IMDb genre 
avg_imdb_rating_by_genre = (
    df.groupby("genre_imdb")["rating_imdb"]
    .mean()
    .reset_index()
    .sort_values("rating_imdb", ascending=False)
)

print("Average IMDb Rating by Genre:")
print(avg_imdb_rating_by_genre)


# 1B. Average popularity by TMDB genre
avg_popularity_by_genre = (
    df.groupby("genre_tmdb")["popularity"]
    .mean()
    .reset_index()
    .sort_values("popularity", ascending=False)
)

print("\nAverage Popularity by TMDB Genre:")
print(avg_popularity_by_genre)


# 1C. Counts per genre per decade 
df["decade"] = (df["release_year"] // 10) * 10

genre_decade_counts = (
    df.groupby(["genre_imdb", "decade"])
    .size()
    .reset_index(name="count")
)

print("\nMovie Counts per Genre per Decade:")
print(genre_decade_counts)
```

```python
import matplotlib.pyplot as plt

#  Ratings vs Release Year 
plt.figure(figsize=(10,5))
df.groupby("release_year")["rating_imdb"].mean().plot()
plt.title("Average IMDb Rating vs Release Year")
plt.xlabel("Year")
plt.ylabel("Average IMDb Rating")
plt.tight_layout()
plt.show()

#  Popularity vs Release Year
plt.figure(figsize=(10,5))
df.groupby("release_year")["popularity"].mean().plot()
plt.title("Average Popularity vs Release Year")
plt.xlabel("Year")
plt.ylabel("Popularity")
plt.tight_layout()
plt.show()

#  Budget Over Time 
plt.figure(figsize=(10,5))
df.groupby("release_year")["budget_in_millions"].mean().plot()
plt.title("Average Budget Over Time")
plt.xlabel("Year")
plt.ylabel("Budget (millions)")
plt.tight_layout()
plt.show()

#  Revenue Over Time 
plt.figure(figsize=(10,5))
df.groupby("release_year")["revenue_in_millions"].mean().plot()
plt.title("Average Revenue Over Time")
plt.xlabel("Year")
plt.ylabel("Revenue (millions)")
plt.tight_layout()
plt.show()
```
